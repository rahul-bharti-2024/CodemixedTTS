```python
import os
import numpy as np
from resemblyzer import VoiceEncoder, preprocess_wav
from tqdm import tqdm
base_dir = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Results_mucs/samples_enhprompts"
test_dir = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/test/wavs"

encoder = VoiceEncoder()

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

similarities = []
cnt=0
for file in tqdm(os.listdir(base_dir)):
    test_path = os.path.join(test_dir, file)
    gen_path = os.path.join(base_dir, file)
    
    if os.path.exists(test_path):
        gt_wav = preprocess_wav(test_path)
        gen_wav = preprocess_wav(gen_path)
        
        embed_gt = encoder.embed_utterance(gt_wav)
        embed_gen = encoder.embed_utterance(gen_wav)
        
        sim_score = cosine_similarity(embed_gen, embed_gt)
        similarities.append(sim_score)
        cnt+=1
    

similarities = np.array(similarities)

print("SSIM Mean:   ", np.mean(similarities))
print("SSIM Median: ", np.median(similarities))
print("SSIM StdDev: ", np.std(similarities))
```

```python
# import librosa
# from dtw import accelerated_dtw  # Install via: pip install dtw-python
# import numpy as np

# def compute_mfcc(path, sr=16000, n_mfcc=13):
#     y, _ = librosa.load(path, sr=sr)
#     mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
#     return mfcc.T  # Shape: (time, features)

# def compute_dtw_mfcc_distance(gt_path, gen_path):
#     mfcc_gt = compute_mfcc(gt_path)
#     mfcc_gen = compute_mfcc(gen_path)
    
#     dist, _, _, _ = accelerated_dtw(mfcc_gt, mfcc_gen, dist='euclidean')
    
#     norm_dist = dist / min(len(mfcc_gt), len(mfcc_gen))  # Normalize by number of frames
#     return norm_dist

# base_dir = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Results_mucs/samples_enhprompts"
# test_dir = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/test/wavs"
# all_distances=[]

# for file in tqdm(os.listdir(base_dir)):
#     test_path = os.path.join(test_dir, file)
#     gen_path = os.path.join(base_dir, file)
    
#     if os.path.exists(test_path):
#         dist= compute_dtw_mfcc_distance(test_path,gen_path)
#         all_distances.append(dist)


# avg_dist = np.mean(all_distances)
# median_dist = np.mean(all_distances)
# std_dev_dist= np.mean(all_distances)

# print("DTW MFCC Mean:   ", avg_dist)
# print("DTW MFCC Median: ",median_dist)
# print("DTW MFCC StdDev: ", std_dev_dist)
import librosa
from fastdtw import fastdtw  # Install via: pip install fastdtw
from scipy.spatial.distance import euclidean
import numpy as np
import os
from tqdm import tqdm


def compute_mfcc(path, sr=16000, n_mfcc=13):
    y, _ = librosa.load(path, sr=sr)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return mfcc.T  

def compute_dtw_mfcc_distance(gt_path, gen_path):
    mfcc_gt = compute_mfcc(gt_path)
    mfcc_gen = compute_mfcc(gen_path)
    
    
    dist, _ = fastdtw(mfcc_gt, mfcc_gen, dist=euclidean)
    

    norm_dist = dist / min(len(mfcc_gt), len(mfcc_gen))
    return norm_dist


base_dir = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Results_mucs/samples_enhprompts"
test_dir = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/test/wavs"


all_distances = []


for file in tqdm(os.listdir(base_dir)):
    test_path = os.path.join(test_dir, file)
    gen_path = os.path.join(base_dir, file)
    
    if os.path.exists(test_path):
        dist = compute_dtw_mfcc_distance(test_path, gen_path)
        all_distances.append(dist)


avg_dist = np.mean(all_distances)
median_dist = np.median(all_distances)  
std_dev_dist = np.std(all_distances)


print("DTW MFCC Mean:   ", avg_dist)
print("DTW MFCC Median: ", median_dist)
print("DTW MFCC StdDev: ", std_dev_dist)
```
