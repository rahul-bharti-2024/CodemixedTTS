```python
import json

input_file = "/home/rahul_b/IndicVoices-R/VoiceCraft/datasets/mucs_roman/manifests/metadata_mucs_test_romanized2.jsonl"
output_file = "/home/rahul_b/IndicVoices-R/VoiceCraft/datasets/mucs_roman/manifests/metadata_mucs_test_romanized3.jsonl"  # Overwriting same file, so we use a temp variable below

# Step 1: Read and parse all JSON objects first
data = []
with open(input_file, 'r', encoding='utf-8') as infile:
    for line in infile:
        if line.strip():  # skip empty lines
            obj = json.loads(line)
            data.append(obj)

# Step 2: Write all JSON objects in pretty format
with open(output_file, 'w', encoding='utf-8') as outfile:
    for obj in data:
        json.dump(obj, outfile, ensure_ascii=False, indent=2)
        outfile.write('\n')
```

```python
import json

input_path = "/home/rahul_b/IndicVoices-R/VoiceCraft/datasets/mucs_roman/manifests/hnyms_metadata.jsonl"
output_file="/home/rahul_b/IndicVoices-R/VoiceCraft/datasets/mucs_roman/manifests/hnyms_metadata1.jsonl"
# Read the file line by line, treating each block as a JSON object
with open(input_path, "r") as infile:
    lines = infile.read().split("\n")
    
# Merge lines into individual JSON blocks
blocks = []
current_block = []
for line in lines:
    if line.strip() == "":
        continue
    current_block.append(line)
    if line.strip().endswith("}"):
        try:
            blocks.append(json.loads("".join(current_block)))
        except Exception as e:
            # print("Error parsing block:", current_block)
            raise e
        current_block = []

# Rewrite as proper JSONL
with open(output_file, "w") as outfile:
    for item in blocks:
        outfile.write(json.dumps(item) + "\n")

```
