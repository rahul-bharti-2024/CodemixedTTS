```python
import pandas as pd
import re
from collections import defaultdict

# Define true punctuation marks you care about
TRUE_PUNCTUATION = set([
    '.', ',', '?', '!', ':', ';', '-', 'â€”', 'â€“', '"', "'", 'â€œ', 'â€',
    'â€˜', 'â€™', '(', ')', '[', ']', '{', '}', 'à¥¤', '...', 'â€¦'
])

# Compile regex pattern to find these punctuations
punct_pattern = re.compile(r"([{}])".format(re.escape(''.join(TRUE_PUNCTUATION))))

# Load your dataset
df = pd.read_csv("/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/train/metadata.csv")

# Store examples where punctuation appears
punct_examples = defaultdict(list)

# Iterate through sentences
for text in df['text'].dropna().astype(str):
    matches = set(punct_pattern.findall(text))
    for match in matches:
        if len(punct_examples[match]) < 5:  # limit to 5 examples per punctuation
            punct_examples[match].append(text)

# Show examples
for punct, examples in punct_examples.items():
    print(f"\nðŸ”¹ Punctuation: '{punct}' â€” {len(examples)} example(s):\n")
    for ex in examples:
        print(f"   â†’ {ex}")
```

```python
import re
def is_devanagari(word):
    """Check if a word contains Devanagari characters."""
    return any('\u0900' <= char <= '\u097F' for char in word)

def normalize_punctuation(sentence):
    # Surround true punctuation with spaces
    sentence = re.sub(r'([à¥¤.,!?;:"â€œâ€â€˜â€™(){}\[\]])', r' \1 ', sentence)
    # Remove multiple spaces
    return ' '.join(sentence.split())
def transliterate_mixed_sentence(sentence):
    sentence = normalize_punctuation(sentence)
    words = sentence.strip().split()
    transliterated_words = []

    for word in words:
        if is_devanagari(word):
            roman = transliterate(word, sanscript.DEVANAGARI, sanscript.ITRANS)
            transliterated_words.append(roman)
        else:
            transliterated_words.append(word)

    return ' '.join(transliterated_words)


import os 
new_csv_test_path="/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/test/metadata_romanized_test.csv"
new_csv_train_path="/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/train/metadata_romanized_train.csv"
og_csv_train_path="/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/train/metadata.csv"
og_csv_test_path="/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/test/metadata.csv"

def process_and_save(input_path, output_path):
    df = pd.read_csv(input_path)
    if 'text' not in df.columns:
        raise ValueError(f"'text' column not found in {input_path}")
    
    df['text'] = df['text'].astype(str).apply(transliterate_mixed_sentence)
    df.to_csv(output_path, index=False)
    print(f"âœ… Saved: {output_path}")

# Run for both train and test
process_and_save(og_csv_train_path, new_csv_train_path)
process_and_save(og_csv_test_path, new_csv_test_path)
```
