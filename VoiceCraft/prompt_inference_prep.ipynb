```python
import os
import json

jsonl_path = "/home/rahul_b/IndicVoices-R/VoiceCraft/datasets/mucs_roman/manifests/metadata_mucs_test_romanized.jsonl"
output_jsonl = "/home/rahul_b/IndicVoices-R/VoiceCraft/datasets/mucs_roman/manifests/metadata_mucs_test_romanized2.jsonl"

prompt_audio = "/mnt/LS226/LS25/rahul2022387/OpenSLR/Dataset/test/wavs/103085_w5Jyq3XMbb3WwiKQ_0002.wav"
prompt_text = "yahA.N hama apane ऑpareTiMga sisTama ke rUpa meM gnu/linux aura libaraऑphisa varjana 334 kA upayoga kara rahe haiM"

with open(jsonl_path, "r", encoding="utf-8") as infile, open(output_jsonl, "w", encoding="utf-8") as outfile:
    for line in infile:
        data = json.loads(line)

        # Use basename of the line's audio_filepath
        output_filename = os.path.basename(data["audio_filepath"])

        # Add prompt-related fields
        data["prompt_audio"] = prompt_audio
        data["prompt_text"] = prompt_text
        data["output_filename"] = output_filename

        # Write updated JSON line
        outfile.write(json.dumps(data, ensure_ascii=False) + "\n")
```
